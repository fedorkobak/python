# \<DataFrame/Series\>.iloc

`iloc` is a way of selecting elements from DataFrames/Series using it index position like this `[<row number>,<column number>]`.

In the next cell I create the data frame for the experiments.

```python
import pandas as pd
import numpy as np
from IPython.display import HTML

sample_size = 10

test_df = pd.DataFrame(
    {
        "a":np.random.choice(range(10), sample_size),
        "b":np.random.choice(range(10), sample_size)
    },
    index = [chr(i) for i in range(ord("a"), ord("a")+sample_size)]
)
test_df
```

# Editable

You can edit element that you access throw `.iloc`.

So in following example I edited one element by accessing it through `iloc`.

```python
edit_df_test = test_df.copy()
display(HTML("<b>Initial df</b>"))
display(edit_df_test)

edit_df_test.iloc[1,1] = "hello"
display(HTML("<b>Edited df</b>"))
display(edit_df_test)
```

**Note** You can only edit an item via `iloc` if you access it directly. If you get a table column via `df.iloc[<row>]` and then apply the `[<column>]` operator to it, it won't work.

So in the following example I try to edit `pandas.DataFrame` using `df.iloc[<row>][<column>]` to edit the dataframe, but find out that the dataframe has not been edited.

```python
edit_df_test = test_df.copy()
display(HTML("<b>Initial df</b>"))
display(edit_df_test)

edit_df_test.iloc[1][1] = "hello"
display(HTML("<b>Edited df</b>"))
display(edit_df_test)
```
